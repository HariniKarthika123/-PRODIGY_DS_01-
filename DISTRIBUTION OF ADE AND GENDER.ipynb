##DISTRIBUTION OF BOTH AGE AND GENDER

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Import necessary libraries
# (Libraries already imported above)

# Step 2: Create a sample dataset (or load your dataset)
np.random.seed(0)
data = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female', 'Other'], 1000, p=[0.48, 0.48, 0.04]),
    'Age': np.random.randint(1, 100, 1000)  # Generate 1000 random ages between 1 and 100
})

# Step 3: Process the dataset
# Count the occurrences of each gender
gender_counts = data['Gender'].value_counts()

# Step 4: Plot the histogram for ages
plt.figure(figsize=(12, 5))

# Histogram for age distribution
plt.subplot(1, 2, 1)
plt.hist(data['Age'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Ages in the Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)

# Step 5: Plot the bar chart for genders
plt.subplot(1, 2, 2)
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink', 'purple'], edgecolor='black', alpha=0.7)
plt.title('Distribution of Genders in the Population')
plt.xlabel('Gender')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
